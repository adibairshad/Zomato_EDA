### Importing libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
plt.style.use('ggplot')
```

### Loading the dataset

```python
df = pd.read_csv('zomato.csv')
df.head()
```

Most of the attributes have object data type, we will need to change few.

```python
#droppping all the columns that are of no use in the analysis
df = df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'], axis = 1)
df.head()
```

```python
df.info()
```

```python
# Null values
df.isnull().sum()
```

```python
#let's look at the duplicates
df.duplicated().sum()
```

```python
#dropping all the duplicates
df.drop_duplicates(inplace = True)
df.shape
```

**Let's look at the rate column first.**

```python
df['rate'].unique()
```

### cleaning the 'rate' column

```python
def clean_rate (value):
    if(value == 'NEW' or value == '-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

df['rate'] = df['rate'].apply(clean_rate)
df['rate'].head()
```

```python
df.isnull().sum()
```

```python
sns.histplot(data=df,x='rate')
```

```python
#as the dist of rate column is almost symmetric we will use mean to fill null values in rate column

df['rate'] = df['rate'].fillna(df['rate'].mean())
df.isnull().sum()
```

```python
#dropping all the other null values
df = df.dropna()
df.isnull().sum()
```

```python
df.columns
```

```python
# renaming the column names to increase readability
df.rename(columns = { 'approx_cost(for two people)' : 'cost_2plates','listed_in(type)' : 'type', 'listed_in(city)'  : 'city'}, inplace = True)
df.head()
```

```python
df['location'].nunique()
```

```python
df['location'].unique()
```

```python
df['city'].nunique()
```

```python
df['city'].unique()
```

```python
# city and location, both are somewhat similar and where going to drop one
df = df.drop(['city'], axis = 1)
```

```python
df['cost_2plates'].unique()
```

```python
#removing comma (,) from cost_2plates column
df['cost_2plates'] = df['cost_2plates'].str.replace(',', '')
```

```python
df['cost_2plates'] = df['cost_2plates'].astype(float)
```

```python
df.dtypes
```

```python
df['rest_type'].value_counts()
```

```python
#we will make a cluster of rest_type with less than 1000 chategory as other
rest_type = df['rest_type'].value_counts(ascending = False)

others = rest_type[rest_type<1000]
others
```

```python
#creating a function to return others in places of values from others

def other(value):
    if(value in others):
        return 'others'
    else:
        return value

df['rest_type'] = df['rest_type'].apply(other)
df['rest_type'].value_counts()
```

```python
df['location'].value_counts()
```

```python
# we are going to do the same thing we did with rest_type

location = df['location'].value_counts(ascending=False)

loc1 = location[location<300]

def loc(value):
    if(value in loc1):
        return 'others'
    else:
        return value

df['location'] = df['location'].apply(loc)
df['location'].value_counts()
```

```python
df['cuisines'].value_counts()
```

```python
cuisines = df['cuisines'].value_counts(ascending=False)

cuisines1 = cuisines[cuisines<100]

def cuisine(value):
    if(value in cuisines1):
        return 'others'
    else:
        return value

df['cuisines'] = df['cuisines'].apply(cuisine)
df['cuisines'].value_counts()
```

### Let's look at the ratings,table booking and online orders

```python
# Create a figure with subplots
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# Plot for Online Order vs Rating
plt.subplot(1, 2, 1)
plt.title('Online Order vs Rating')
sns.boxplot(data=df, x='online_order', y='rate',palette="pastel")
plt.xlabel('Online Order')
plt.ylabel('Rating')

# Plot for Book Table vs Rating
plt.subplot(1, 2, 2)
plt.title('Book Table vs Rating')
sns.boxplot(data=df, x='book_table', y='rate', palette="pastel")
plt.xlabel('Book Table')
plt.ylabel('Rating')

# Adjust layout
plt.tight_layout()

plt.show()
```

**There are outliers in the data. But we can see that the median rating for both online order yes and no is almost same.
While for book table the median rating is more in case of table booking availability than non availability.**

We can conclude that people are giving more rating to restaurant with booking availabilty.

```python
#let's group online order and location, then create a pivot table
df1 = df.groupby(['location','online_order'])['name'].count()
df1.to_csv('location_online.csv')
df1 = pd.read_csv('location_online.csv')
df1 = pd.pivot_table(df1, values= None, index=['location'], columns=['online_order'],fill_value=0, aggfunc=np.sum)
df1.head()
```

```python
# Filter out 'Others' category from df1
filtered_df1 = df1[df1.index != 'others']

plt.figure(figsize=(15, 8))
plt.title("Location with online availability")
filtered_df1.plot(kind='bar', ax=plt.gca())  # Plotting on the current axis
plt.xlabel('Location')
plt.ylabel('')
plt.xticks(rotation=90)
plt.show()
```

**The location with most restaurants BTM, max restaurants provide online delivery.**

```python
df2 = df.groupby(['location','book_table'])['name'].count()
df2.to_csv('location_booktable.csv')
df2 = pd.read_csv('location_booktable.csv')
df2 = pd.pivot_table(df2,values = None,index=['location'], columns=['book_table'], fill_value=0, aggfunc=np.sum)
df2.head()
```
